```python
import matplotlib.pyplot as plt
import os
import json
import re
```

```python
log_path = r"models\cad-diffusion-128px\training.log"
with open(log_path, 'r') as f:
    log = json.load(f)
```

```python
pattern = r"total_iteration (\d+), epoch (\d+), step (\d+) \| loss ([\d\.e\-]+) \| lr: ([\d\.e\-]+)"
losses = []
for line_dict in log:
    line : str = line_dict['data']
    if line.startswith("total_iteration"):
        match = re.search(pattern, line)
        total_iteration = int(match.group(1))
        epoch = int(match.group(2))
        step = int(match.group(3))
        loss = float(match.group(4))
        lr = float(match.group(5))
        
        losses.append(loss)
```

```python
plt.plot(losses)
plt.title("cat-diffosion-128px")
plt.xlabel('step')
plt.ylabel('loss')
plt.show()
```
